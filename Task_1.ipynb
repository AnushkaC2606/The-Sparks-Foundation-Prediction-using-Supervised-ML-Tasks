```python
# Step1 importing the libraries

import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
from sklearn import linear_model 
```

```python
#Read the required database from csv file given

data = pd.read_csv("http://bit.ly/w-data")
data.head(10)
```

```python
#Printing a plot and visualized the data w.r.t to x and y axis

data.plot(kind='scatter',x='Hours', y='Scores', style='o')  
plt.title('No. of hrs vs marks scored')
plt.show()
```

# Fitting/Training Model



```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values 
```

```python
#Spliting the data into training and testing set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=50) 
```

```python

reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
```

```python
#Plotting the Regression line

plt.scatter(X, y)
plt.plot(X, line,color='r')
plt.show()

line = reg.coef_ * X + reg.intercept_
```

```python
#Predict the data

print(X_test) # Testing data - In Hours
y_predict = reg.predict(X_test) # Predicting the scores
```

```python
#Compare the actual and predicted Values

data = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict}) 
data
```

```python
print("training score", reg.score(X_train,y_train))
print("testing score", reg.score(X_test,y_test))
```

```python
# What will be predicted score if a student studies for 9.25 hrs/ day?

hours = 9.25
own_predict = reg.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_predict[0]))
```

Evaluating the model

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_predict)) 
```
